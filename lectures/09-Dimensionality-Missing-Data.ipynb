## Exercise 1

Apply PCA with 2 dimensions to the iris dataset, then plot the resulting projections.

```python
import csv
import numpy as np
X = []
y = []
with open('../data/datasets/iris/iris.csv') as in_file:
    iCSV = csv.reader(in_file, delimiter=',')
    for row in iCSV:
        y.append(row[-1])
        X.append([float(x) for x in row[:-1]])
X = np.array(X)
y = np.array(y)
```

```python
# Write code to transform data using PCA here
# Save transformed data into a matrix called "X_small". You can delete the following line of code.
X_small = np.random.random((X.shape[0], 2))
```

```python
%matplotlib inline

# DO NOT Change this cell!!!

import matplotlib
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = y == target
    ax.scatter(X_small[indicesToKeep,0]
               , X_small[indicesToKeep,1]
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
